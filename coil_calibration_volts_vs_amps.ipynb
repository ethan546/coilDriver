```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as p
from scipy.stats import stats
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
from scipy.optimize import curve_fit
import sympy as sp
```

Here are the calibration curves for inout voltage to transistors vs. current through coils (logarithm curve). Each coil has two curves corresponding to data taken while magnetic field was measured in both the positive and negative directions. csv files containing data can be found in Benchling, posted on 3/9/20.

### Black Coils

```python
#pandas reads csv files, skiprows to remove unnecessary data and outliers
f1 = p.read_csv('Black_coil_calibration_positive.csv', skiprows=[1,2])
f2 = p.read_csv('Black_coil_calibration_minus.csv', skiprows=[1,2])
```

```python
#uncomment to view data tables
#f1
#f2
```

```python
#Magnetic field measured in plus direction

#plot data
plt.plot(f1.amps, f1.Volts, 'o', label = 'Data')
plt.ylabel("Voltage input to Arduino (V)")
plt.xlabel('Current through coils (A)')


#fit an exponential
def test(x, a, b, c): 
    return a*np.log(b *(c + x))
param, param_cov = curve_fit(test, f1.amps, f1.Volts, p0 = [3.84615,3.77283,0.28]) 
#print('optimized parameters =', param)
#print('covariance matrix =', param_cov)
print('fitted function:', param[0],'* log(',param[1],'* ( x +', param[2],'))')
plt.plot(f1.amps, test(f1.amps, *param), 'r', label = 'Fitted Exponential')

plt.legend(loc='upper left')
plt.title("Black Coil: Volts vs. Amps")
```

```python
#Magnetic field measured in minus direction

#plot data
plt.plot(f2.amps, f2.volts, 'o', label = 'Data')
plt.ylabel("Voltage input to Arduino (V)")
plt.xlabel('Current through coils (A)')


#fit and plot an exponential
def test(x, a, b, c,): 
    return a*np.log(b *(c + x))
param, param_cov = curve_fit(test, f2.amps, f2.volts, p0 = [3.84615,3.77283,0.28]) 
#print('optimized parameters =', param)
#print('covariance matrix =', param_cov)
print('fitted function:', param[0],'* log(',param[1],'* x +', param[2],')')
plt.plot(f2.amps, test(f2.amps, *param), 'r', label = 'Fitted Exponential')

plt.legend(loc='upper left')
plt.title("Black Coil: Volts vs. Amps ")
```
